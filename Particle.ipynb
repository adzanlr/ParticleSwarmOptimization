```python
import java.util.Random;

public class Particle {
    int Ss;
    int Ph;
    int Pn;
    int k;
    int l;
    int s;
    int b;
    int Ph1;
    int Pn1;
    int k1;
    int l1;
    int s1;
    int Ss1;
    int Ph2;
    int Pn2;
    int k2;
    int l2;
    int s2;
    int b1;
    int Ss2;

    public Particle(
            String[] Makanansusunih,
            String[] Proteinhewani,
            String[] Proteinnabati,
            String[] Karbohidrat,
            String[] Lemakaja,
            String[] Sayur,
            String[] Buahaja,
            boolean isFirstIteration
    ) {
        if (isFirstIteration) {
            Ss = new Random().nextInt(Makanansusunih.length);
            Ph = new Random().nextInt(Proteinhewani.length);
            Pn = new Random().nextInt(Proteinnabati.length);
            k = new Random().nextInt(Karbohidrat.length);
            l = new Random().nextInt(Lemakaja.length);
            s = new Random().nextInt(Sayur.length);
            b = new Random().nextInt(Buahaja.length);
            Ph1 = new Random().nextInt(Proteinhewani.length);
            Pn1 = new Random().nextInt(Proteinnabati.length);
            k1 = new Random().nextInt(Karbohidrat.length);
            l1 = new Random().nextInt(Lemakaja.length);
            s1 = new Random().nextInt(Sayur.length);
            Ss1 = new Random().nextInt(Makanansusunih.length);
            Ph2 = new Random().nextInt(Proteinhewani.length);
            Pn2 = new Random().nextInt(Proteinnabati.length);
            k2 = new Random().nextInt(Karbohidrat.length);
            l2 = new Random().nextInt(Lemakaja.length);
            s2 = new Random().nextInt(Sayur.length);
            b1 = new Random().nextInt(Buahaja.length);
            Ss2 = new Random().nextInt(Makanansusunih.length);
        }
    }

    public void setAll(
        int _Ss,
        int _Ph,
        int _Pn,
        int _k,
        int _l,
        int _s,
        int _b,
        int _Ph1,
        int _Pn1,
        int _k1,
        int _l1,
        int _s1,
        int _Ss1,
        int _Ph2,
        int _Pn2,
        int _k2,
        int _l2,
        int _s2,
        int _b1,
        int _Ss2
    ) {
        Ss = _Ss;
        Ph = _Ph;
        Pn = _Pn;
        k = _k;
        l = _l;
        s = _s;
        b = _b;
        Ph1 = _Ph1;
        Pn1 = _Pn1;
        k1 = _k1;
        l1 = _l1;
        s1 = _s1;
        Ss1 = _Ss1;
        Ph2 = _Ph2;
        Pn2 = _Pn2;
        k2 = _k2;
        l2 = _l2;
        s2 = _s2;
        b1 = _b1;
        Ss2 = _Ss2;
    }

    public double[] getXKal(
        double[] kalorisusu,
        double[] kalorihewani,
        double[] kalorinabati,
        double[] karbokalori,
        double[] kalorilemak,
        double[] kalorisayur,
        double[] kaloribuah
    ) {
        return new double[] { kalorisusu[Ss], kalorihewani[Ph], kalorinabati[Pn], karbokalori[k], kalorilemak[l], kalorisayur[s], kaloribuah[b],
                kalorihewani[Ph1], kalorinabati[Pn1], karbokalori[k1], kalorilemak[l1], kalorisayur[s1], kalorisusu[Ss1], kalorihewani[Ph2],
                kalorinabati[Pn2], karbokalori[k2], kalorilemak[l2], kalorisayur[s2], kaloribuah[b1], kalorisusu[Ss2] };
    }

    public double[] getXPro(
        double[] proteinhewani,
        double[] proteinnabati
    ) {
        return new double[] { proteinhewani[Ph], proteinhewani[Ph1], proteinhewani[Ph2], proteinnabati[Pn], proteinnabati[Pn1], proteinnabati[Pn2] };
    }

    public double[] getXKar(
            double[] karbokarbo
    ) {
        return new double[] { karbokarbo[k], karbokarbo[k1], karbokarbo[k2] };
    }

    public double[] getXLem(
            double[] lemaklemak
    ) {
        return new double[] { lemaklemak[l], lemaklemak[l1], lemaklemak[l2] };
    }
}
```
